Since our dataset is very **imbalanced**, presenting a much higher quantity of non-donors than donors, it is necessary to perform a **balancing of the data** in order to avoid overfitting the models to the overrepresented class. This balancing can be done by either over or under-sampling. Both approaches were tested in this work and we concluded that undersampling yields much worse results than its alternative. However, when trying to balance a dataset by oversampling, we must take into account the effect of the generated samples on the cross validation step of our models. If we were to perform the oversampling during the pre-processing part of the work, later when dividing the dataset into train and test splits for cross validation, some of the generated samples would be assigned to the test group, while the rest would be assigned into the training group. Since these samples are procedurally generated from other samples on the dataset, this would lead to having similar rows of data between the training and the test groups. That could cause **overfitting** of the models to the data on these rows during the training step and therefore lead to over optimistic estimates of model performance when classifying similar rows on the test group. 

In order to avoid this issue, we implemented the oversampling technique inside of the cross validation step, making sure that only the training set is oversampled and there are **no repeated rows between groups**.

Besides that, there are multiple possible ways to oversample a dataset. The most simple one is by using the **RandomOverSampler**, that picks random rows from the minority group and creates copies of them until the number of samples between groups are even. However, there are more complex ways of oversampling a dataset, for example, using the **SMOTE** or **ADASYN** techniques. 

What **SMOTE** does is simple. First it finds the n-nearest neighbors in the minority class for each of the samples, then it draws a line between the neighbors and generates random points on the lines.

**ADASYN** can be considered an improved version of **SMOTE**. It does the same as SMOTE, but with a minor improvement. After generating the samples it adds a random small value to the points making it more realistic, by adding a little variance in order to avoid a linear correlation between the generated points.

Due to the availability of different oversampling techiques, we created a function that tests the three of them given our classifier of choice and outputs a plot with the **ROC curves** for each oversampling method, using a given supervised ML model. 


```python
def test_classif(classifier,splits=10):
    
    LW = 2
    RANDOM_STATE = 1
    
    cv = StratifiedKFold(n_splits=splits)

    classifier = ['Model', classifier]

    samplers = [
        ['ADASYN', ADASYN(random_state=RANDOM_STATE)],
        ['ROS', RandomOverSampler(random_state=RANDOM_STATE)],
        ['SMOTE', SMOTE(random_state=RANDOM_STATE)],
    ]

    pipelines = [
        ['{}-{}'.format(sampler[0], classifier[0]),
         make_pipeline(sampler[1], classifier[1])]
        for sampler in samplers
    ]

    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)

    for name, pipeline in pipelines:
        mean_tpr = 0.0
        mean_fpr = np.linspace(0, 1, 100)
        for train, test in cv.split(X, y):
            probas_ = pipeline.fit(X[train], y[train]).predict_proba(X[test])
            fpr, tpr, thresholds = roc_curve(y[test], probas_[:, 1])
            mean_tpr += np.interp(mean_fpr, fpr, tpr)
            mean_tpr[0] = 0.0
            roc_auc = auc(fpr, tpr)

        mean_tpr /= cv.get_n_splits(X, y)
        mean_tpr[-1] = 1.0
        mean_auc = auc(mean_fpr, mean_tpr)
        plt.plot(mean_fpr, mean_tpr, linestyle='--',
                 label='{} (area = %0.2f)'.format(name) % mean_auc, lw=LW)

    plt.plot([0, 1], [0, 1], linestyle='--', lw=LW, color='k',
             label='Luck')

    # make nice plotting
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    ax.spines['left'].set_position(('outward', 10))
    ax.spines['bottom'].set_position(('outward', 10))
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')

    plt.legend(loc="lower right")

    plt.show()
```

We ran the oversampling tests on 4 different supervised learning models: **KNN**, **Logistic Regression**, **Gaussian Nayve-Bayes** and **Decision Tree Classifier**.

```python
test_classif(KNeighborsClassifier(n_neighbors=5, n_jobs=-1))
```

```python
test_classif(LogisticRegression(n_jobs=-1))
```

```python
test_classif(GaussianNB())
```

```python
test_classif(DecisionTreeClassifier(max_depth=4))
```

By looking at the **Area Under Curve (AUC)** values for the different oversampling techniques, we come to the conclusion that in our case study, the basic oversampling algorithm (**RandomOverSampler**), always performs as well or even better than the more complex methods. For this reason, we decided to use this oversampling technique for the rest of this work.

Having selected the oversampling method to use, we then create a function to test supervised learning models by performing cross-validation and displaying as output the corresponding **Accuracy** and **Recall Score** for each model.

```python
from sklearn import metrics

def test_model(X, y, model, folds = 10, recall_avg = "binary"):

    cv = StratifiedKFold(n_splits=folds)
    sum_acc = []
    sum_recall = []
    
    for train_idx, test_idx, in cv.split(X, y):
        X_train, y_train = X[train_idx], y[train_idx]
        X_test, y_test = X[test_idx], y[test_idx]
        X_train, y_train = RandomOverSampler().fit_sample(X_train, y_train)
        rfc=model
        if (type(model).__name__) == "RIPPER":
            rfc.fit(X_train, y_train, pos_class = 1.0)
            sum_acc.append(rfc.score(X_test, y_test, metrics.accuracy_score))
            sum_recall.append(rfc.score(X_test, y_test, metrics.recall_score))
        else:
            rfc.fit(X_train, y_train)
            preds = rfc.predict(X_test)
            sum_acc.append(metrics.accuracy_score(y_test, preds))
            sum_recall.append(metrics.recall_score(y_test, preds, average = recall_avg))
    
    output = "Accuracy: "+ str(np.mean(sum_acc)) + " (" + str(np.std(sum_acc)) + ") " + "Recall: "+ str(np.mean(sum_recall)) + " (" + str(np.std(sum_recall)) + ")\n"
    print(output)
```

As we can see, the Rule model - **RIPPER** - presents the highest accuracy, however its recall score is very low. The **Logistic Regression** on the other hand presents the second lowest accuracy score, but the highest recall score.

Since we are dealing with a naturally imbalanced dataset, we found the recall score to be the best metric when choosing the best performing model, since the accuracy scores can be easily inflated by the presence of a high number of false positives. Recall is also often regarded in literature as the best metric to use when there is an uneven class distribution and it's important to reduce the number of **false negatives**. 

For this reason, we decided to choose the **Logistic Regression** as our best model to predict **TARGET_B** and therefore perform its subsequent optimization by tuning the model hyperparameters.

In order to find the best parameters to **optimize for the recall score** of the chosen model, we used a function that makes use of the **IMB Learn Pipeline** module to perform a **grid search** on the parameters, while still performing the oversampling during cross validation for any given model.  

```python
def Imb_GridSearchCV(X, y, model, parameters, balance, folds = 10, scorer = "accuracy", recall_avg = "binary"):
    
    #Define the score used for GridSearchCV to find the best params
    from sklearn.metrics import make_scorer, recall_score, accuracy_score
    
    if scorer == "accuracy":
        scorer = make_scorer(accuracy_score)
    elif scorer == "recall":
        scorer = make_scorer(recall_score, average = recall_avg)
    
    # Using the ImbLearn pipeline allows to oversample just the training sets
    from imblearn.pipeline import Pipeline

    pipeline = Pipeline([
            ('balance', balance),
            ('model', model) ])

    #Change the keys od dictionary to have model__ for the pipeline
    hyperparameters = parameters.copy()
    for key in parameters.keys():
        hyperparameters['model__'+key] = hyperparameters.pop(key)
    
    
    # Use GridSearchCV 
    from sklearn.model_selection import GridSearchCV
    clf = GridSearchCV(pipeline, hyperparameters, n_jobs = -1, cv=folds, scoring = scorer)

    #Fit the model to all the dataset in order to get the best parameters
    best_model = clf.fit(X,y)
    scores = best_model.best_score_

    
    # output the best_model and its parameters
    
    return best_model.best_estimator_[1], best_model.best_estimator_.get_params()
```

Here, we create a dictionary of Logistic Regression parameters for the Grid Search function to use and test the results of the different combinations of parameters using a 10-fold cross validation.

```python
parameters = {'penalty' : ['l1', 'l2'],
              'C' : [0.001, 0.01, 0.1, 1, 10, 100, 1000],
              'solver' : ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}

best_params = Imb_GridSearchCV(X, y, LogisticRegression(n_jobs=-1), parameters, RandomOverSampler(), scorer = "recall",  folds = 10)[0]
print(best_params)
```

The result of this Grid Search indicates that the best parameters to fit the Logistic Regression model to our data are: 
- C = 0.01
- penalty = l1
- solver = liblinear

The "C" parameter represents the inverse of regularization strength (which favors simpler prediction
rules to avoid overfitting). Small C values imply a stronger regularization.

"penalty" is the penalization norm used against the logistic model for having too many variables. Because our solver parameter is liblinear, the penalty parameter cannot be set with the option "none".

The "solver" parameter represents the algorithm that is going to be used in the optimization problem.

```python
test_model(X, y, best_params, folds=10, recall_avg = "binary")
```

We considered that **False Negatives would be particularly problematic** in our case study, since it is **very important  to avoid discarding potential donors**, while sending solicitations to individuals that end up not donating wouldn't be as problematic. This can be achieved by looking at the recall estimate, that gives us information about **'how many relevant items are selected'** in a binary classification by dividing the number of true positives by the number of true positives + false negatives.

The reason why we didn't try to maximize accuracy was that when in the presence of naturally skewed binary targets we must avoid models that maximize only for this metric, since those models would have the tendency to classify almost all samples as non-donors. This would achieve high accuracy values on our test sets since the distribution of the target variable is skewed in that direction, but wouldn't be interesting in terms of provided insights for the charitable organization.

As such, by maximizing for **recall**, we make sure that the charitable organization messages as much potential donors as possible, even if the selection of message recipients is not always accurate. This is important because the cost of missing an important donor would far outweigh the cost of messaging non-donating individuals.

For this reason we can maximize for any of these parameters and obtain the same results.

### Association Rules
Rule generation is a common task in the mining of frequent patterns. An association rule is an implication expression of the form X→Y, where X and Y are disjoint itemsets. There are several metrics used to evaluate an association rule:
* **Support** - the support of the combined itemset A**∪**C (where A is the antecedent itemset and C is the consequent itemset)
* **Confidence** - the probability of the consequent in a transaction given that it also contains the antecedent.
* **Lift** - measure how much more often the antecedent and consequent of a rule A->C occur together than if they were statistically independent. 
* **Leverage** - the difference between the observed frequency of A and C appearing together and the frequency that would be expected if A and C were independent.
* **Conviction** - measures how much the consequent is dependent on the antecedent.

To more easily search for rules, we defined the following function, that helps with the filtering of the obtained rules through **association_rules *from* mlxtend.frequent_patterns**.

```python
# Function that filters association rules with the given parameters
def rules_filtered(fi, min_antecedent_len=1, min_consequent_len=1, max_antecedent_len=10, max_consequent_len=10, min_support=None, min_confidence=0, min_conviction=None, max_conviction=None, min_leverage=None):
    # Get rules with min_confidence
    rules = association_rules(fi, metric="confidence", min_threshold=min_confidence)
    
    # Filter antecedent and consequent length
    rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
    rules["consequent_len"] = rules["consequents"].apply(lambda x: len(x))
    rules.drop(rules[rules['antecedent_len'] > max_antecedent_len].index, inplace = True)
    rules.drop(rules[rules['consequent_len'] > max_consequent_len].index, inplace = True)
    rules.drop(rules[rules['antecedent_len'] < min_antecedent_len].index, inplace = True)
    rules.drop(rules[rules['consequent_len'] < min_consequent_len].index, inplace = True)
    
    # Filter with min_support
    if(min_support!=None):
        rules.drop(rules[rules['support'] < min_support].index, inplace = True)
    
    # Filter with min_conviction
    if(min_conviction!=None):
        rules.drop(rules[rules['conviction'] < min_conviction].index, inplace = True)
    # Filter with max_conviction
    if(max_conviction!=None):
        rules.drop(rules[rules['conviction'] > max_conviction].index, inplace = True)
            
    # Filter with min_leverage
    if(min_leverage!=None):
        rules.drop(rules[rules['leverage'] < min_leverage].index, inplace = True)
        
    
    rules.reset_index(drop=True, inplace=True)
    return rules
```

## 2.5. Finding Groups

#### KMeans algorithm

This algorithm starts by randomly choosing a 'k' number of centroids for a 'k' number of clusters, where 'k' is initially defined by the user. Then, it attributes each dataset observation to the nearest cluster (nearest centroid). When all observations are assigned to a cluster, it adjusts the centroid positions of the clusters according to the complete set of observations assigned to each of them. It then repeats the steps of assigning all the observations to the new controids and shift again the positions of the centroids. The process stops when the centroids reach an optimal position within the cluster and no more centroid positional changes are recorded.

To apply this algorithm to our dataset we start by creating a new dataframe dfKM, identical to our pre processed dataframe df. 

Note that this dataframe does not include the feature TARGET_D, as we have dropped it in the preprocessing stage. This allows us to make cluster predictions without knowing the amount of money that the individuals have donated to the organization, and consequently which donation type category they belong to.

```python
dfKM = pd.DataFrame(df)
dfKM
```

Here, we define the classifier km for this algorithm, specifying the 'k' number of clusters that we aim to obtain with our input dataset. We choose to divide the dataset in 5 clusters as we want to test if donors belonging to each of the 5 different donation type categories (A, B, C, D and E) can potentially be separated in these 5 clusters. This would show that individuals who donate similar amounts of money would also have similar personal and behavioral data, collected throughout the different features of our dataframe. The "random_state" allow us to specify a seed for this run of the algorithm and therefore obtain the same result of predicted clusters even if we run it multiple times, which facilitates the presentation and discussion of the results.

In addition to fitting the km classifier to our dataframe dfKM and predict the cluster to which each donor likely belongs to, we  assign this array of predicted clusters to the variable "predicted". The lenght of "predicted" is equal to the number of donors, which is also the number of rows in our dataframe dfKM. 
We then include this array as a new column in dfKM, which we name "CLUSTER".

The first lines of the dataframe and the column with the predicted groups/clusters are displayed below.

```python
km = KMeans(n_clusters=5, random_state=12)

predicted = km.fit_predict(dfKM)

dfKM['CLUSTER'] = predicted

dfKM.head()
```

We build the matrix using the columns "DONATION_TYPE" (corresponding to the Donation type categories) and "CLUSTER" (corresponding to the predicted cluster), and assign it to the variable cmKM (contingency matrix of the KMeans clustering).

#### Agglomerative Clustering algorithm

The Agglomerative Clustering algorithm starts by defining every instance of the dataset as an independent cluster, and based on the distance metrics considered and the threshold defined, begins to merge the clusters that are nearest to each other in an hierarchical way, until there are only 'k' clusters left, where 'k' is set by the user.

Again, we start by creating a new dataframe dfHAC (dataframe for Hierarchical Agglomerative Clustering), which is identical to our previously pre-processed dataframe df.

```python
dfHAC = pd.DataFrame(df)
```

We define the classifier hac (hierarchical agglomerative clustering) for this algorithm, and the number of clusters is again set to 5. 
Additionally, we have to define the linkage parameter, which determines the type of distance metrics that are considered when linking (or not) the different groups of clusters. We choose to assign this parameter with the ward linkage method, which minimizes the variance between the clusters that are merged.
Consequently, when the ward option is chosen for the linkage, the affinity parameter (metric used to compute the linkage) has obligatorily to be set to euclidean distances.

We then fit the agglomerative clustering classifier to our dataframe dfHAC and predict the cluster of each donor, assigning the array of cluster predictions to the variable 'predicted'. Again, we append this array to our dataframe as a new column called 'CLUSTER'.

The first lines of the dataframe and the column with the predicted groups/clusters are displayed below.

```python
hac = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')

predicted = hac.fit_predict(dfHAC)

dfHAC['CLUSTER'] = predicted

dfHAC.head()
```
